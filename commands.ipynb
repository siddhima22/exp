```python
#FM Algo

nums = list(map(int, input("Enter numbers separated by space: ").split()))

def h1(x): return (6 * x + 1) % 5

def trailing_zeros(n):
    b = bin(n)[2:]                  
    return len(b) - len(b.rstrip('0'))  

hashed = [h1(x) for x in nums]

R = max([trailing_zeros(h) for h in hashed]) if hashed else 0

r = 2 ** R

print("Hashed values:", hashed)
print("Estimated number of distinct elements:", r)
```

```python
#Bloom

size = 10
bloom = [0] * size

def h1(x):return (3 * x + 7) % size

def h2(x):return (5 * x + 11) % size

while True:
    ch = input("\n1.Insert  2.Query  3.Show  4.Exit : ")

    if ch == "1":
        x = int(input("Enter number: "))
        bloom[h1(x)] = 1
        bloom[h2(x)] = 1
        print("Inserted!")

    elif ch == "2":
        x = int(input("Enter number: "))
        if bloom[h1(x)] and bloom[h2(x)]:
            print(f"{x}  May be present  ")
        else:
            print(f"{x} Definitely not present ")

    elif ch == "3":
        print("Bits:", bloom)

    elif ch == "4":
        break

    else:
        print("Invalid choice!")
```
